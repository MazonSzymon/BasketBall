```
#Import libraries
import pandas as pd
import numpy as np
```

```
# Import dataset
url = 'https://raw.githubusercontent.com/MazonSzymon/BasketBall/master/players_stats_by_season_full_details.csv'
orginal_data = pd.read_csv(url)
```

```
orginal_data.head()
```

```
orginal_data.info()
```

```
df = orginal_data.copy()
df.columns = ['League Name','Season Year','Stage',
                        'Player Full Name','Team name', 'Games Played',
                        'Minutes Played', 'Field Goals Made', 'Field Goals Attempts',
                        '3 Points Made', '3 Points Attemps', 'Free Throws Made',
                        'Free Throws Attempts','Turnovers','Personal Fouls',
                        'Offensive Rebounds','Defensive Rebounds','Rebounds',
                        'Assists','Steals','Blocks',
                        'Points','Birth Year','Births Month',
                        'Birth Date', 'Height (Feet)', 'Height (CM)',
                        'Weight (Pounds)','Weight (KG)','Nationality',
                        'High School']
df.head()
```

```
df.info(memory_usage='deep')
```

```
# Change columns data type
#df['League Name'].value_counts() It should be category type
df['League Name'] = df['League Name'].astype('category')
#df['Season Year'].value_counts() It should be a category type
df['Season Year'] = df['Season Year'].astype('category')
#df['Stage'].value_counts() It should be a category type
df['Stage'] = df['Stage'].astype('category')
# df['Team name'].value_counts() It could be category type
df['Team name'] = df['Team name'].astype('category')
df['Birth Year'] = df['Birth Year'].fillna(0).astype('int')
#df['Nationality'].value_counts() It could be a category type
df['Nationality'] = df['Nationality'].astype('category')
```

```
df.info(memory_usage='deep')
```

```
# Delete some not intresting info
df.drop(columns= ['Birth Year','Birth Date','Height (Feet)','Weight (Pounds)'], inplace=True)
```
